```python
from keras.layers import Conv2D,BatchNormalization,Activation
import numpy as np
from keras.models import Sequential

def resBlock(X,f):
  X_copy=X;

  X=Conv2D(f,kernel_size=(1,1),strides=(1,1),kernel_initializer='he_uniform')(X)
  X=BatchNormalization()(X)
  X=Activation('relu')(X)

  X=Conv2D(f,kernel_size=(3,3),padding='same',strides=(1,1),kernel_initializer='he_uniform')(X)
  X=BatchNormalization()(X)

  X_copy=Conv2D(f,kernel_size=(1,1),strides=(1,1),kernel_initializer='he_uniform')(X_copy)
  X_copy=BatchNormalization()(X_copy)

  X=Add()(X,X_copy)
  X=Activation('relu')(X)
```

```python
from keras.layers import UpSampling2D,Concatenate

def upsampling2D(x,skip):
  x=UpSampling2D((2,2))(x)
  merge=Concatenate()([x,skip])
  return merge
```

```python
from keras.layers import Input, Conv2D, BatchNormalization, Activation, MaxPooling2D, UpSampling2D, Add
from keras.models import Model

input_shape = (256,256,3)

# Input tensor shape
X_input = Input(input_shape)

# Stage 1
conv1_in = Conv2D(16,3,activation= 'relu', padding = 'same', kernel_initializer ='he_normal')(X_input)
conv1_in = BatchNormalization()(conv1_in)
conv1_in = Conv2D(16,3,activation= 'relu', padding = 'same', kernel_initializer ='he_normal')(conv1_in)
conv1_in = BatchNormalization()(conv1_in)
pool_1 = MaxPooling2D(pool_size = (2,2))(conv1_in)

# Stage 2
conv2_in = resBlock(pool_1, 32)
pool_2 = MaxPooling2D(pool_size = (2,2))(conv2_in)

# Stage 3
conv3_in = resBlock(pool_2, 64)
pool_3 = MaxPooling2D(pool_size = (2,2))(conv3_in)

# Stage 4
conv4_in = resBlock(pool_3, 128)
pool_4 = MaxPooling2D(pool_size = (2,2))(conv4_in)

# Stage 5 (Bottle Neck)
conv5_in = resBlock(pool_4, 256)

# Upscale stage 1
up_1 = upsampling2D(conv5_in, conv4_in)
up_1 = resBlock(up_1, 128)

# Upscale stage 2
up_2 = upsampling2D(up_1, conv3_in)
up_2 = resBlock(up_2, 64)

# Upscale stage 3
up_3 = upsampling2D(up_2, conv2_in)
up_3 = resBlock(up_3, 32)

# Upscale stage 4
up_4 = upsampling2D(up_3, conv1_in)
up_4 = resBlock(up_4, 16)

# Final Output
output = Conv2D(1, (1,1), padding = "same", activation = "sigmoid")(up_4)

model = Model(inputs = X_input, outputs = output )

model.summary()
```
